```python
def get_stock_api_dummies(df_original):
    # create an empty dataframe
    df_converted=pd.DataFrame()
    
    # loop through all rows in the original dataframe
    for ind in df_original.index: 
        
        # another loop to go through all items in the ticker column
        for ticker in df_original["tickers"][ind]:
            
            #create a temporary list and append it to the new dataframe
            col_list=[[df_original["date"][ind],df_original["title"][ind],df_original["source_name"][ind],df_original["sentiment"][ind],ticker]]
            df_converted=df_converted.append(col_list)
    
    # rename columns
    df_converted.columns=['date','title','source_name','sentiment','ticker']
    
    # create dummies and assign to the converted dataframe itself
    df_converted=pd.get_dummies(df_converted, columns=['ticker'])
    return df_converted
```
